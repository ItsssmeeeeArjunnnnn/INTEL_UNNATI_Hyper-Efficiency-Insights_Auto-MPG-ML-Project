### Convert Notebook to HTML

To convert this Colab notebook (`.ipynb`) to an HTML file, we can use `nbconvert`. First, we'll install it, then use the command line tool to perform the conversion.

```python
# Install nbconvert
!pip install nbconvert
```

```python
import nbformat
from nbconvert.exporters import HTMLExporter
import os

notebook_name = '/content/drive/MyDrive/Colab Notebooks/INTEL_UNNATI_Hyper-Efficiency Insights_Auto-MPG ML Project' # Removed .ipynb extension
output_html_name = '/content/drive/MyDrive/Colab Notebooks/INTEL_UNNATI_Hyper-Efficiency Insights_Auto-MPG ML Project.html'

# Ensure the output directory exists
output_dir = os.path.dirname(output_html_name)
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load the notebook
with open(notebook_name, 'r', encoding='utf-8') as f:
    notebook_content = nbformat.read(f, as_version=4)

# Configure and export to HTML
exporter = HTMLExporter()
output, resources = exporter.from_notebook_node(notebook_content)

# Write the output HTML to a file
with open(output_html_name, 'w', encoding='utf-8') as f:
    f.write(output)

print(f"Successfully converted '{notebook_name}' to '{output_html_name}'.")
print("You can find the generated HTML file in the Colab file browser (folder icon on the left panel). Or, in Google Drive at the same location as your notebook.")
```
